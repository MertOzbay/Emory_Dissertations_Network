```python
import little_mallet_wrapper
import seaborn
import glob
from pathlib import Path
import numpy as np

path_to_mallet = '../../mallet-2.0.8/bin/mallet'
```

```python
num_topics = 25

output_directory_path = 'topic-model-output'
path_to_training_data           = f"{output_directory_path}/training.txt"
path_to_formatted_training_data = f"{output_directory_path}/mallet.training"
path_to_model                   = f"{output_directory_path}/mallet.model.{str(num_topics)}"
path_to_topic_keys              = f"{output_directory_path}/mallet.topic_keys.{str(num_topics)}"
path_to_topic_distributions     = f"{output_directory_path}/mallet.topic_distributions.{str(num_topics)}"
path_to_diagnostics             = f"{output_directory_path}/diagnostics.xml"
path_to_word_weights            = f"{output_directory_path}/word.weights"

topic_dist_list = little_mallet_wrapper.load_topic_distributions(path_to_topic_distributions)
```

```python
print(len(theses_df))
print(len(topic_dist_list))
```

```python
dep_2_topic_vectors = dict()
dep_2_topic_vectors.clear()

for idx, row in theses_df.iterrows():
    dist = topic_dist_list[idx]
    dep = row['department']
    if not (type(dep) == str): continue
        
    if dep not in dep_2_topic_vectors:
        dep_2_topic_vectors[dep] = [0]*25
    old_dist = dep_2_topic_vectors[dep] 
    for topic_num, topic in enumerate(old_dist):
        old_dist[topic_num] = dist[topic_num] + old_dist[topic_num]
    dep_2_topic_vectors[dep] = old_dist
    
for dep in dep_2_topic_vectors:
#     topic_list = dep_topic_vectors[dep]
    dep_2_topic_vectors[dep] = [topic / dep_count[dep] for topic in dep_2_topic_vectors[dep]]
#     dep_topic_vectors[dep]/len(dep_topic_vectors[dep])

dep_2_topic_vectors
```

```python
for dep in dep_2_topic_vectors:
    print(dep + ": \t\t"+str(sum(dep_2_topic_vectors[dep])))
```

```python
for dep in dep_2_topic_vectors:
    print("Dep name: " + dep)
    print(dep_2_topic_vectors[dep])
    print()
```

```python
topic_by_dep = []

for i in range(0,25):
    
    topic_by_dep.append(dict()) 
    for dep in dep_2_topic_vectors:
        topic_by_dep[i][dep] = dep_2_topic_vectors[dep][i]

topic_by_dep
```

```python
topics = little_mallet_wrapper.load_topic_keys(path_to_topic_keys)

for topic_number, topic in enumerate(topics):
    print(f"✨Topic {topic_number}✨\n\n{topic}\n")
    top_deps = sorted(topic_by_dep[topic_number].items(), key=lambda x: x[1], reverse=True)[:5]
    for dep in top_deps:
        print(str(dep[1])+ ":\t" + dep[0])
```
